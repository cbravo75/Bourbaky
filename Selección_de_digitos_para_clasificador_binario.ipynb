## Digitos

```python
from sklearn.datasets import fetch_openml
mnist = fetch_openml('mnist_784')
mnist.keys() #to get keys
```

```python
X = mnist['data']
print(type(X))
X.shape
```

```python
randDigit = X[32]
randDigitImage = randDigit.reshape(28,28)
plt.imshow(randDigitImage, cmap='Greys')
plt.axis("off")
plt.show()
```

```python
def showDigits(instances, images_per_row=20):
    images_per_row = min(len(instances), images_per_row)
    images = [instance.reshape(28,28) for instance in instances]
    n_rows = (len(instances) - 1) // images_per_row + 1
    row_images = []
    n_empty = n_rows * images_per_row - len(instances)
    images.append(np.zeros((28, 28 * n_empty)))
    for row in range(n_rows):
        rimages = images[row * images_per_row : (row + 1) * images_per_row]
        row_images.append(np.concatenate(rimages, axis=1))
    image = np.concatenate(row_images, axis=0)
    plt.imshow(image, cmap = 'Greys')
    plt.axis("off")
```

```python
plt.figure(figsize=(10,5))
example_images = X[:105]
showDigits(example_images, 15)
plt.show()
```
